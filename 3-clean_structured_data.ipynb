```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("structured_data.csv", sep = ";")
```

```python
# clean 

data["temperature"] = np.where((data["temperature"] < 47) & (data["temperature"] > 30) , data["temperature"], np.nan)
data["bloodPressure"] = np.where((data["bloodPressure"] > 50) & (data["bloodPressure"] < 300) , data["bloodPressure"], np.nan)
data["admitFrom"] = np.where(data["admitFrom"] != "Eigen woonomgeving" , data["admitFrom"], "Home")
data["admitFrom"] = np.where(data["admitFrom"] != "Revalidatie instelling" , data["admitFrom"], "Nursing or Rehabilitation")
data["admitFrom"] = np.where(data["admitFrom"] != "Verpleeg-/verzorgingshuis" , data["admitFrom"], "Nursing or Rehabilitation")
data["admitFrom"] = np.where(data["admitFrom"] != "Ziekenhuis, algemeen" , data["admitFrom"], "Hospital")
data["admitFrom"] = np.where(data["admitFrom"] != "Ziekenhuis, ander academisch ziekenhuis" , data["admitFrom"], "Hospital")
data["admitFrom"] = np.where(data["admitFrom"] != "Ziekenhuis, eigen Poli" , data["admitFrom"], "Hospital")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Eigen woonomgeving" , data["dischargeTo"], "Home")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Overleden met obductie" , data["dischargeTo"], "Died")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Overleden zonder obductie" , data["dischargeTo"], "Died")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Revalidatiecentrum" , data["dischargeTo"], "Rehabilitation centre")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Verpleeg-/verzorgingshuis" , data["dischargeTo"], "Nursing home")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Ziekenhuis, algemeen" , data["dischargeTo"], "Hospital")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Ziekenhuis, ander academisch ziekenhuis" , data["dischargeTo"], "Hospital")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Overgeplaatst naar militair hospitaal, noodhospitaal of ziekenboeg van een penitentiaire inrichting" , data["dischargeTo"], "Hospital")
data["dischargeTo"] = np.where(data["dischargeTo"] != "Tegen medisch advies in vertrokken" , data["dischargeTo"], "Left against medical advice")
data = data.set_index("id")
data["BMI"] = data.groupby("id").mean()["BMI"]
data["BMI"] = np.where(data["BMI"].isna(), 28.6, data["BMI"])
data = data.reset_index()
```

```python
filled_temperature = pd.DataFrame() 

for pseudo_id in data["id"].drop_duplicates():
    data1 = data[data["id"] == pseudo_id]
    data1["date"] = pd.to_datetime(data1["date"])
    data2 = data1.set_index("date")["temperature"].interpolate(method="linear")
    data2 = data2.reset_index()
    i = 0
    while i < len(data1) and pd.isna(data2.iloc[i,1]):
        i += 1
    if i >= len(data1):
        data2["temperature"] = 36.8
    else:
        for j in range(i):
            data2.iloc[j,1] = data2.iloc[i,1] 
    data2["id"] = pseudo_id
    filled_temperature = filled_temperature.append(data2)
    
filled_blood_pressure = pd.DataFrame() 

for pseudo_id in data["id"].drop_duplicates():
    data1 = data[data["id"] == pseudo_id]
    data1["date"] = pd.to_datetime(data1["date"])
    data2 = data1.set_index("date")["bloodPressure"].interpolate(method="linear")
    data2 = data2.reset_index()
    i = 0
    while i < len(data1) and pd.isna(data2.iloc[i,1]):
        i += 1
    if i >= len(data1):
        data2["bloodPressure"] = 85
    else:
        for j in range(i):
            data2.iloc[j,1] = data2.iloc[i,1] 
    data2["id"] = pseudo_id
    filled_blood_pressure = filled_blood_pressure.append(data2)
    
filled_o2Saturation = pd.DataFrame() 

for pseudo_id in data["id"].drop_duplicates():
    data1 = data[data["id"] == pseudo_id]
    data1["date"] = pd.to_datetime(data1["date"])
    data2 = data1.set_index("date")["o2Saturation"].interpolate(method="linear")
    data2 = data2.reset_index()
    i = 0
    while i < len(data1) and pd.isna(data2.iloc[i,1]):
        i += 1
    if i >= len(data1):
        data2["o2Saturation"] = 94.5
    else:
        for j in range(i):
            data2.iloc[j,1] = data2.iloc[i,1] 
    data2["id"] = pseudo_id
    filled_o2Saturation = filled_o2Saturation.append(data2)
```

```python
data = data.set_index("id")
filled_o2Saturation = filled_o2Saturation.set_index("id")
data["o2Saturation"] = filled_o2Saturation["o2Saturation"]
filled_blood_pressure = filled_blood_pressure.set_index("id")
data["bloodPressure"] = filled_blood_pressure["bloodPressure"]
filled_temperature = filled_temperature.set_index("id")
data["temperature"] = filled_temperature["temperature"]
```

```python
data.to_csv("cleaned_structured_data.csv", sep=";")
```
